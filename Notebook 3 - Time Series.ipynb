```python
data['2012'].resample('D').plot()
```

```python
#monthly mean and median concentration of the 'FR04037' station for the years 2009-2012 
data.loc['2009':,'FR04037'].resample('M',how='mean').plot()
data.loc['2009':,'FR04037'].resample('M',how='median').plot()
```

```python
data.loc['2009':, 'FR04037'].resample('M', how=['mean', 'median']).plot()
```

```python
#QUESTION: plot the monthly mininum and maximum daily concentration of the 'BETR801' station 
daily = data['BETR801'].resample('D')
daily.resample('M', how=['min', 'max']).plot()
```

```python
#Bar plot of the mean of stations in year 2012
data['2012'].mean().plot(kind='bar')
```

```python
#QUESTION: The evolution of the yearly averages with, and the overall mean of all stations?
data.resample('A').plot()
data.mean(axis=1).resample('A').plot(color='k', linestyle='--', linewidth=4)
```

```python
#Typical monthly profile for different stations
#Add a column to the dataframe to indicate the month
data['month'] = data.index.month
```

```python
data.groupby('month').mean()
```

```python
#Plot monthly profile for different stations
data.groupby('month').mean().plot()
```

```python
#Weekly 95 percentile of concentration in stations 'BETR801' and 'BETN029' for 2011 
df2011 = data['2011']
df2011.groupby(df2011.index.week)[['BETN029','BETR801']].quantile(0.95).plot()
```
